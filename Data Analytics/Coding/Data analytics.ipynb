```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
PATH_DATA = '../Data/'
```

```python
df_del = pd.read_csv(PATH_DATA+'food_delivery_datasets.csv')
```

```python
df_del
```

```python
a = df_del[['resto_id', 'status']]
a
```

```python
df_del['date_time'] = pd.to_datetime(df_del['date_time']).dt.round('H')
df_del['date_time']
```

```python
hari = df_del['date_time'].dt.strftime('%A')
hari
```

```python
jam = df_del['date_time'].dt.strftime('%H:%M')
jam
```

```python
df_new = pd.concat([a['resto_id'], hari, jam, a['status']], axis=1)
df_new.columns = ['resto_id', 'days', 'time', 'status']
df_new
```

```python
grup_status = df_new.groupby('status')
grup_status['status'].size()
```

```python
df = df_new[df_new.status == 'Completed']
df.reset_index(drop=True)
```

```python
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
plot1 = df.groupby(['days']).size().reindex(days).plot(kind='bar', figsize = (10,5), ylabel='freq', 
title='Frequencies towards days of the period')
```

```python
plot1.figure.savefig('../plot/frekuensi per hari.jpg')
```

```python
df.groupby(['days']).size().reindex(days)
```

```python
fig, ax = plt.subplots()
hours = mdates.HourLocator(interval = 1)

hm = df.groupby(['time']).size()

ax.plot(hm.index, hm.values)
hm.plot(ax = ax, x_compat=True, figsize = (10,6), ylabel='freq', 
        title='Frequencies towards times of the period')
ax.xaxis.set_major_locator(hours)
fig.autofmt_xdate()
```

```python
fig.savefig('../plot/frekuensi per waktu periode.jpg')
```

```python
group_days = df.groupby(['days', 'time']).size().unstack(level=0, fill_value=0)
plot3 = group_days.plot(figsize=(10,6), ylabel='freq',
                        title='Multigraph of frequencies towards times of the period')
```

```python
plot3.figure.savefig('../plot/multigraf frekuensi per waktu periode.jpg')
```

```python
df.groupby(['days', 'time']).size().idxmin()
```

```python
fig = plt.figure('scatter dates', figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

grp_days = df.groupby(['days', 'time'], as_index=False).size()
time = grp_days['time'].unique()
time.sort()

grp_days['time'] = grp_days['time'].replace(time, [i for i in range(len(time))])
grp_days['days'] = grp_days['days'].replace(days, [i for i in range(len(days))])

x = grp_days['days']
y = grp_days['time']
z = grp_days['size']

plot = ax.scatter(x, y, z, c=z)
fig.colorbar(plot)

ax.xaxis.set_ticklabels(days)
ax.set_yticks([i for i in range(0,len(time),3)])
ax.yaxis.set_ticklabels([time[i] for i in range(0,len(time),3)])

plt.show()
```
